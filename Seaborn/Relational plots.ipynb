```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

## Relplot

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="day", col="time", data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="day", col="time", row="sex", data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="time", col="day", col_wrap=2, data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="time", size="size", palette=["b", "r"], sizes=(10, 100), \
                col="time", data=tips)
```

```python
fmri = sns.load_dataset("fmri")
g = sns.relplot(x="timepoint", y="signal",hue="event", style="event", 
                col="region",kind="line", data=fmri)
```

```python
 g = sns.relplot(x="timepoint", y="signal", hue="event", style="event", col="region",
                 height=5, aspect=.7, kind="line", data=fmri)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", palette=cmap, data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 200),
                    palette=cmap,legend="full", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 200),
                    hue_norm=(0, 7),legend="full", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", palette="Set2", data=tips)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
ax = sns.scatterplot(x="total_bill", y="tip", style="time", markers=markers,data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", s=100, color=".2", marker="+", data=tips)
```

```python
iris = sns.load_dataset("iris")
ax = sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species, style=iris.species)
```

```python
index = pd.date_range("1 1 2000", periods=100, freq="m", name="date")
data = np.random.randn(100, 4).cumsum(axis=0)
wide_df = pd.DataFrame(data, index, ["a", "b", "c", "d"])
ax = sns.scatterplot(data=wide_df)
```
